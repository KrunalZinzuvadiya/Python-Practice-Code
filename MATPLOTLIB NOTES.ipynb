## Matplotlib Pie Charts

With Pyplot, you can use the pie() function to draw pie charts.

A simple pie chart:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()

#   By default the plotting of the first wedge starts from the x-axis and moves counterclockwise
```

```python
# Adding labels to the pie plot

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
# EXPLODE : Taking out a specific portion from pie plot

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]   # pulling the wedge 'Apple' by 0.2

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# Adding percentages in pie plot

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, autopct = '%0.2f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
### alpha used to control transparency of the markers

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(0,100,100)
y = np.random.randint(0,100,100)
colors = np.random.randint(0,100,100)
sizes = 10 * np.random.randint(0,100,100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
